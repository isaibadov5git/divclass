```python
def square(reqem:int):
    kvadrat = reqem**2
    return kvadrat

square(3.5)

def salam_De(adamlar: list):
    salam_cumlesi = 'salam '
    for adam in adamlar:
        salam_cumlesi = salam_cumlesi + f"{adam}, "

    return salam_cumlesi

salam_De(["Ali","isa"])

siyahi = [1,5,10]

sum=0
for reqem in siyahi:
    sum += reqem

print(sum)
```

#### Scope

#### Exercise 10

```python
def find_long_words(words:list, min_length:int):
    new_words=[]
    for word in words:
        if len(word) > min_length:
            new_words.append(word)

    return new_words

        
find_long_words(words=["apple","banana","cherry","date","elderberry"],min_length=5)
```
