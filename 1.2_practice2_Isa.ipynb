```python
reqemler = [-2,-1,1,3,4,7,11]
kvadratlar = [i**2 for i in reqemler]

kvadratlar2 = []
for reqem in reqemler:
    kvadrat = reqem**2
    kvadratlar2.append(kvadrat)

kvadratlar2

```

```python
reqemler = [-2,-1,1,3,4,7,11]
kvadratlar2 = [reqem if reqem%2 == 0 else reqem**2 for reqem in reqemler if reqem > 0] # right side = filter, left side = condition
kvadratlar2

```
