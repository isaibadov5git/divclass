#### Functional programming concepts

```python
def kvadrata_yukselt(x):
    return x**2

cavab = map(kvadrata_yukselt, [1,5,7,10])
list(cavab)
```

```python
list(map(lambda x:x**2, [1,2,3,4,5]))
```

```python
# filter(boolien functin, iterable)
cavab = filter(lambda x:x>=0,[-2,-3,5,-7,0,9]) # same as list comprehension

next(cavab)
```

```python
from functools import reduce

# reduce(function, collection)
def vur(x,y):
    return x*y
reduce(vur, [1,2,3,4,5])
```

### Comprehension

```python
adlar = ["hesen","huseyn","ali","lale","gulnar"]
s = "hesen huseyn ali lale gulnar"
adlar = s.split(" ")
adlar
```

```python
print([ad[0] for ad in adlar])
print([ad[0] for ad in adlar if ad[0] == "h"])
```

```python
list = []
for s in range(-5,5):
    if s%2==0:
        if s > 0:
            kv = s**2
            list.append(kv)
        else:
            list.append(s)

list
```

```python
[(s**2 if s>0 else s) for s in range(-5,5) if s%2==0 ] # filter
```

```python
{ad[0] for ad in adlar} # set comprehension
{ad:ad[0] for ad in adlar}
```

```python
tuple(ad[0] for ad in adlar)
```

```python
def factorial(num):
    pr = 1
    for i in range(1,num+1):
        pr = pr * i

    return pr

{i:factorial(i) for i in range(1,6)}
```

```python
def calculate_discount(price, discount_percent):
    assert discount_percent > 0 and discount_percent < 100, "Invalid discount percentage"

    return price*(100-discount_percent)/100

calculate_discount(100,20)

```
